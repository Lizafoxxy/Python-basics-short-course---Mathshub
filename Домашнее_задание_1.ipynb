### 1. Делаем настоящую анкету пользователя
С клавиатуры вводится имя, фамилия, пол, возраст, род деятельности.
Выведите такую форму
Имя:

Фамилия:

Пол:

Возраст:

Род деятельности:


```python
name = input("Введите Ваше имя: ")
surname = input("Введите Вашу фамилию: ")
gender = input("Укажите Ваш пол: ")
age = int(input("Введите Ваш возраст: "))
job = input("Укажите Ваш род деятельности: ")
print(f"Имя: {name}")
print(f"Фамилия: {surname}")
print(f"Пол: {gender}")
print(f"Возраст: {age}")
print(f"Род деятельности: {job}")
```

### 5. У вас есть три строки, содержащие улицу, номер дома и номер квартиры. Объедините их в полный адрес с разделителями-запятыми и пробелами: "Улица, Дом, Квартира".


```python
a = "Улица Горького"
b = "Дом 5"
c = "Квартира 21"
print(a+",", b+",", c)
print(f"{a}, {b}, {c}")
```

### 6.  Объявление цены товара:
   У вас есть переменная с ценой товара, например 350 рублей. Создайте строку, которая отображает цену с валютой: "Цена товара: 350 рублей".

```python
a = 350
print("Цена товара:", a, "рублей")
print(f"Цена товара: {a} рублей")
```

### 7.Регистрация времени окончания задания:
   Ученик завершил тест в 14 часов и 35 минут. Составьте строку, отражающую время: "Задание выполнено в 14:35".

```python
a = 14
b = 35
print("Задание выполнено в ", a, ":", b, sep="")
print(f"Задание выполнено в {a}:{b}")
```

### 8.  Отобразить текущий год в сообщении:
   Вам дано число, представляющее текущий год - 2023. Сформируйте строку, которая выглядит так: "Текущий год - 2023".

```python
a = 2023
print("Текущий год", a, sep="-")
print("Текущий год-", a, sep="")
print(f"Текущий год-{a}")
```
